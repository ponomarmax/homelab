# 00 Session Discovery

Discover available wearable sessions for a user from remote `raw` + `processed` trees.

```python
from pathlib import Path
import logging
import os
import sys

import pandas as pd
from dotenv import load_dotenv

NOTEBOOK_ROOT = Path.cwd()
LIB_PATH = NOTEBOOK_ROOT / "lib"
if str(LIB_PATH) not in sys.path:
    sys.path.insert(0, str(LIB_PATH))

logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s %(name)s %(message)s")
load_dotenv(NOTEBOOK_ROOT / ".env")

from session_discovery import get_latest_session, list_sessions
```

```python
# Parameters
user_id = 2
session_id = None  # Set explicit session id to pin selection
```

```python
sessions = list_sessions(user_id=user_id, env_path=NOTEBOOK_ROOT / ".env")
print(f"Found {len(sessions)} sessions for user_id={user_id}")

rows = []
for row in sessions:
    rows.append({
        "session_id": row["session_id"],
        "streams": ",".join(row.get("available_streams", [])),
        "artifacts": ",".join(row.get("available_artifacts", [])),
        "dates": ",".join(row.get("dates", [])),
    })

pd.DataFrame(rows)
```

```python
if session_id:
    selected = next((s for s in sessions if s["session_id"] == session_id), None)
    if selected is None:
        raise ValueError(f"session_id={session_id} not found")
else:
    selected = get_latest_session(user_id=user_id, env_path=NOTEBOOK_ROOT / ".env")

print("Selected session:")
print(f"  session_id: {selected['session_id']}")
print(f"  streams: {selected.get('available_streams', [])}")
print(f"  dates: {selected.get('dates', [])}")
print(f"  raw paths: {len(selected.get('raw_paths', []))}")
print(f"  processed paths: {len(selected.get('processed_paths', []))}")
```
